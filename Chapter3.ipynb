```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

Binomial distribution

```python
n=10
p=0.3
mu, var = binom.stats(n,p,moments='mv')
k = np.linspace(0,n,n+1,endpoint=True)
B = binom.pmf(k, n, p)
plt.bar(k,B)
plt.show()
norm = np.sum(B)
mu_num = np.sum(k*B)
var_num = np.sum((k-mu_num)**2*B)
print("norm", norm)
print("mu:", mu, "mu_num:", mu_num)
print("var:", var, "var_num", var_num)
```

Random walkers

```python
import random
p = 0.4
n = 100
m=50
moves = np.random.choice((0,1),size = (n,m),p = (0.6,0.4))
time = np.linspace(0,n,n+1,endpoint=True)
position = np.insert(np.cumsum(moves,axis=0),0,0,axis =0)
plt.plot(time,position)
plt.show()
```

Many random walkers, histograms

```python
from IPython.display import clear_output
import time
fig, ax = plt.subplots()
m = 1000000
bins = np.linspace(-0.5,59.5,60)
moves2 = np.random.choice((0,1),size = (n,m),p = (0.6,0.4))
position2 = np.insert(np.cumsum(moves2,axis=0),0,0,axis =0)
for i in range(n):
  plt.figure(1)
  plt.hist(position2[i,:],bins=bins)
  plt.xlim(-1,56)
  plt.show()
  time.sleep(0.25)
  clear_output(wait=True)
```
